# 0. Example by default

```python
#load ETC with a config file and the COLIBRI caracteristics
falcon_etc=etc(configFile="falcon_test.hjson",
        name_telescope='falcon')

# Execute
falcon_etc.sim()
```

# 1. Compute the SNR for a given magnitude and exposure time

We keep using COLIBRI telescope and let’s say we want to compute the SNR reached for a magnitude of 18 (AB) in a single exposure of 20s in the i band.

```python
# It will also display information and plots but we do not show them here. If you just want the SNR, you can get it through:

falcon_etc.information['SNR']
falcon_etc.information['etc_type'] = 'snr'
falcon_etc.information['object_type'] = 'magnitude'
falcon_etc.information['object_magnitude'] = 11.208
falcon_etc.information['exptime'] = 60
falcon_etc.information['filter_band'] = 'V'
falcon_etc.sim()
```

```python
# It will also display information and plots but we do not show them here. If you just want the SNR, you can get it through:
falcon_etc.information['SNR']
```

# 2. Compute the limiting magnitude for a given SNR and exposure time

We keep using COLIBRI telescope and this time we want to compute the limiting magnitude in the J band for a SNR of 10 and a single exposure of 30s. We need to use the NIR channel, which is named CAGIRE in the hjson file.

```python
falcon_etc.information['etc_type'] = 'mag'
falcon_etc.information['SNR'] = 900
falcon_etc.information['exptime'] = 20
falcon_etc.information['channel'] = 'APX1' #Esto está en el example
falcon_etc.information['filter_band'] = 'V' #Esto está en el example

falcon_etc.sim()
```

# 1. Exposure Time Calculator

```python
# If you just want the exposure time, you can get it through:

#Seeing in arcsec at 500nm and for an airmass 1 
falcon_etc.information["seeing_zenith"] = 1
falcon_etc.information["elevation"] = 50.0
falcon_etc.information["moon_age"] = 4.76
falcon_etc.information["Temp_night"] = 5

falcon_etc.information['etc_type'] = 'time'
falcon_etc.information['object_type'] = 'magnitude'
falcon_etc.information['object_magnitude'] = 11.208
falcon_etc.information['filter_band'] = 'V'
falcon_etc.information['SNR'] = 150
falcon_etc.sim()
```
